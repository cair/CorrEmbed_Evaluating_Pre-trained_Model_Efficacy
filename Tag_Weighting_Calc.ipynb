```python
def group_outfit_tags_by_category(tag_df, outfit_tag_df, outfit_id):
    outfit_tags = tag_df[tag_df["id"].isin(outfit_tag_df[outfit_tag_df["outfitsId"] == outfit_id]["tagsId"])]
    tags_by_category = {}
    for row in outfit_tags.iterrows():
        if row[1]["tagCategory"] not in tags_by_category:
            tags_by_category[row[1]["tagCategory"]] = []
        tags_by_category[row[1]["tagCategory"]].append(row[1]["tag"])
    return tags_by_category


tag_df.groupby("tagCategory").head()
#group_outfit_tags_by_category(tag_df, outfit_tag_df, "outfit.6e2a3eeffbf4433c9d6e5085d8c083f2")
rented_outfits_by_user["outfit_tags"] = rented_outfits_by_user["outfit"].apply(lambda x: group_outfit_tags_by_category(tag_df, outfit_tag_df, x))
rented_outfits_by_user
```

```python
import numpy as np
import copy

def calculate_tag_similarity_score(user_outfit_tags):
    tag_dict = {tag_cat_df[column]["category"]:{tag:0 for tag in list(tag_cat_df[column]["tags"]) + ["None"]} for column in tag_cat_df}
    for outfit_tag_dict in user_outfit_tags["outfit_tags"].values:
        for cat, tags in tag_dict.items():
            if cat not in outfit_tag_dict:
                tag_dict[cat]["None"] += 1
                continue
            for outfit_tag in outfit_tag_dict[cat]:
                tag_dict[cat][outfit_tag] += 1
    return tag_dict

def calculate_tag_probabilites(tag_dict, outfit_count):
    tag_probabilities = copy.deepcopy(tag_dict)
    for column in tag_cat_df:
        cat_name = tag_cat_df[column]["category"]
        for tag in tag_probabilities[cat_name]:
            tag_probabilities[cat_name][tag] = tag_probabilities[cat_name][tag] / outfit_count
    return tag_probabilities

def calculate_entropy(probabilites):
    entropy = 0
    for prob in probabilites:
        if prob == 0:
            continue
        #prob = prob if prob != 0 else 0.000001
        entropy += prob * np.log(prob)
    entropy *= -1
    return entropy

def calculate_tag_entropy(tag_probabilites):
    tag_entropy = {}
    for cat, tags in tag_probabilites.items():
        #del tags["None"]
        tag_values = tags.values()
        entropy_lower = 0#calculate_entropy([ 1 if i == 0 else 0 for i in range(len(tag_values))])
        # TODO: Max entropy in this case might simply be 1 per value rather than the mean value. 
        entropy_upper = calculate_entropy([1/len(tag_values) for i in range(len(tag_values))])
        entropy = calculate_entropy(tag_values)
        #print(f"Category bounds {cat}: {entropy / entropy_upper} - {entropy_lower} - {entropy_upper} - {entropy}")
        tag_entropy[cat] = {"entropy":entropy ,"normalized":entropy / entropy_upper, "lower":entropy_lower, "upper":entropy_upper}
        
    return tag_entropy

def calculate_tag_weighting_per_user(user_id):
    user_outfits = rented_outfits_by_user[rented_outfits_by_user["renter"] == user_id]
    outfit_counts = len(user_outfits)
    tag_occurences = calculate_tag_similarity_score(user_outfits)
    tag_cat_probabilites = calculate_tag_probabilites(tag_occurences, outfit_counts)
    tag_cat_entropy = calculate_tag_entropy(tag_cat_probabilites)
    return tag_cat_entropy, tag_cat_probabilites


#rented_outfits_by_user.iloc[4]["outfit_tags"]
recurring_renters_df = pd.DataFrame(recurring_renters).reset_index()
#set column names
recurring_renters_df.columns = ["renter", "rental_count"]
arb_user_id = recurring_renters_df["renter"].iloc[0]
entropy_scores, occurence_dict = calculate_tag_weighting_per_user(arb_user_id)
entropy_scores
```

```python
def merge_dicts(dict_list):
    merged_dict = {key:[value] for key, value in dict_list[0].items()}
    for to_merge in dict_list[1:]:
        for key, value in to_merge.items():
            merged_dict[key].append(value)
    for key, scores in merged_dict.items():
        merged_dict[key] = np.mean(scores)
    return merged_dict

average_entropy = {}
for renter in recurring_renters_df["renter"].values:
    entropy_scores, occurence_dict = calculate_tag_weighting_per_user(renter)
    for cat, score in entropy_scores.items():
        if cat not in average_entropy:
            average_entropy[cat] = []
        average_entropy[cat].append(score)
average_entropy = {cat:merge_dicts(scores) for cat, scores in average_entropy.items()}
entropy_df = pd.DataFrame(average_entropy)
entropy_df
```

```python
def parse_user_weights(tag_dict):
    tag_weight_attributes = {}
    for cat in tag_dict.keys():
        total_occurences = sum(tag_dict[cat].values())
        # if total_occurences < 7:
        #     continue
        tag_percentage = []
        tags = []
        for tag in tag_dict[cat].keys():
            portion_of_cat = tag_dict[cat][tag] / total_occurences
            tag_percentage.append(portion_of_cat)
            tags.append(tag)
            #print(tag, portion_of_cat)
        tag_weight_attributes[f"{cat}_sum"] = total_occurences
        tag_weight_attributes[f"{cat}_max"] = max(tag_percentage)
        tag_weight_attributes[f"{cat}_min"] = min(tag_percentage)
        tag_weight_attributes[f"{cat}_all"] = tag_percentage
        tag_weight_attributes[f"{cat}_all_tags"] = tags
        tag_weight_attributes[f"{cat}_max_tag"] = tags[tag_percentage.index(max(tag_percentage))]

        #print(max(tag_percentage), min(tag_percentage), sum(tag_percentage) / len(tag_percentage))
        #print(cat, list(tag_dict[cat].keys())[tag_percentage.index(max(tag_percentage))], total_occurences)
    return tag_weight_attributes

tag_attributes = []
for user in recurring_renters_df["renter"].values:
    tag_weights = calculate_tag_weighting_per_user(user)
    #attributes = parse_user_weights(tag_weights)
    attributes = parse_user_weights(tag_weights[0])
    attributes["user"] = user
    tag_attributes.append(attributes)


tag_attributes_df = pd.DataFrame.from_dict(tag_attributes)    
```

```python
import re

def cat_weight(cat, ta_df):
    return ta_df[ta_df[f"{cat}_sum"] > 20][f"{cat}_max"].mean()

def cat_sum(cat_sum, ta_df):
    return ta_df[ta_df[cat_sum] > 20][cat_sum].mean()

def parse_cat_name(cat_name):
    re_pattern = "^(.+)_max$"
    if re.search(re_pattern, cat_name):
        return True, re.search(re_pattern, cat_name).group(1)
    return False, None


#tag_attributes_df.columns[["_max" in column for column in tag_attributes_df.columns]]
category_weights = []
for column in tag_attributes_df.columns:
    is_cat_max, cat_name = parse_cat_name(column)
    if is_cat_max:
        category_weights.append({"category": cat_name, "weight": cat_weight(cat_name, tag_attributes_df), "sum": cat_sum(f"{cat_name}_sum", tag_attributes_df)})
weights_df = pd.DataFrame.from_dict(category_weights)
weights_df
```
